# Definitions

```python
theta_max  = 115
theta_min  = 5
theta_range = theta_max - theta_min
omega_fast = -theta_range / (0.6)
omega_slow = theta_range / (3.4)
```

```python
def Omega(theta, t):
    """Return the shell velocity at a given time.
    This generates slightly different results than Raphael's scheme, particularly
    near the closing-opening transition.
    """

    t = t % 4.0
    
    if (t <= 0.02):
        p = t/0.02
        omega = (1.0 - p)*omega_slow + p*omega_fast
        
    elif (t < 0.6):
        omega = omega_fast
    
    elif( t <= 0.62):
        p = (t-0.6)/0.02
        omega = (1.0 - p)*omega_fast + p*omega_slow
    
    elif(t > 0.62):
        omega = omega_slow

    return omega
```

```python
def Theta(omega, t):
    """Generate shell angles"""
    
    return cumtrapz(omega, t, initial=0) + theta_max
```

```python
def COM(head_posn, shell_angle):
    """
    Calculate the centre of mass position.
    
    More than half the mass of the scallop is in the shells, which move relative to the head.
    """

    #Measurements taken from Images.0034
    m_head   = 495
    m_shells = 538.8
    l_shell =  0.20 
    scale = l_shell/(107.75/2)
    hinge_offset = 6.96 * scale
    
    com = (m_head * head_posn + \
           m_shells * (head_posn + hinge_offset + 0.5* l_shell*np.cos(shell_angle*np.pi/180.)))/(m_head + m_shells)        
    return com    
```

```python
def moving_average(a, n=3) :

    data=np.concatenate(([a[0] for i in range(int(np.floor(n/2.)-1))],
                        a,
                        [a[-1] for i in range(int(np.ceil(n/2.)))]))
    ret = np.cumsum(data, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
def draw_vlines(ax):
    for (p,c) in ((3.4, '#FFB280'), (4.0, '#CCE680'), (7.4, '#FFB280'), (8, '#CCE680'), (11.4, '#FFB280'), (12, '#CCE680')):    
        ax.axvline(p, color=c)
```

# Load and Process Data

```python
%cd /home/rafa/work/benchmarkData/Swimmer/fischer/double_checks/final/revision/revision_plot/correspondance/items

# load the y position data from file
data1=np.loadtxt('y-pos.txt')

t = data1[:,0]
head = data1[:,1]

# Load shell angle data from file
data2=np.loadtxt('angles.txt')
t2 = data2[:,0]

# repeat angle data
theta = np.tile(180*data2[:-11,1], 4)

# Load the net force data from file
data3=np.loadtxt('y_force_cycle.txt')
t3 = data3[:,0]

# repeat force data
# the data extends for more than one period, so repeat only the trailing part
F = data3[:,1]
F = np.concatenate((F[:50,], np.tile(F[50:], 3)))
```

```python
# Trim datasets to a common length and shift to that indices are synchronized
# Assume that time values are synchronized between datasets (??) and shift indices to match
# There's a lot of guesswork here.

shift = -int((t[0]-t2[0])/0.0002)
N = 60000
t = t[shift: N+shift]
head = head[shift: N+shift]
theta = theta[:N]

shift = int((t[0]-t3[0]))/0.0002
F = F[shift: N+shift]
```

```python
#Calculate COM
com = COM(head, theta)
```

```python
# Calculate velocity and acceleration using a sparser time grid
spacing = 100
t_s = t[::spacing]
dt = t_s[1]-t_s[0]
y = moving_average(com, 20)[::spacing]
v = moving_average(np.append(np.diff(y), 0)/dt, 5)
a = moving_average(np.append(np.diff(v), 0)/dt, 5)

F_s = moving_average(F, spacing)[::spacing]
```

### Velocity and Acceleration

```python
f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(t_s, v)
axarr[0].set_ylabel('Velocity (dm/s)')
axarr[0].axhline(0, color='k')
draw_vlines(axarr[0])

axarr[1].plot(t_s, a, 'r')
axarr[1].set_ylabel('Acceleration (dm/s^2)')
axarr[1].axhline(0, color='k')
axarr[1].set_xlabel('time (s)')
draw_vlines(axarr[1])
```

```python
py.iplot_mpl(f, strip_style = True)
```

Surprisingly, the largest acceleration is seen at the transition between opening and closing (not the other way around), this may be because of the polygonal shape of the angle profile for this transition.
